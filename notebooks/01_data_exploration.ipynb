```python
import requests
import os

# Direct download from OPUS — no HuggingFace needed
url = "https://object.pouta.csc.fi/OPUS-bible-uedin/v1/moses/am-en.txt.zip"

print("Downloading Amharic-English Bible parallel corpus from OPUS...")
response = requests.get(url, stream=True)

with open("am-en.zip", "wb") as f:
    for chunk in response.iter_content(chunk_size=8192):
        f.write(chunk)

print(f"✓ Downloaded: {os.path.getsize('am-en.zip') / 1024 / 1024:.1f} MB")

# Unzip it
import zipfile
with zipfile.ZipFile("am-en.zip", "r") as zip_ref:
    zip_ref.extractall("data/raw")
    print(f"✓ Extracted files: {zip_ref.namelist()}")
```

```python
# Read both files
with open("data/raw/bible-uedin.am-en.am", "r", encoding="utf-8") as f:
    amharic_lines = f.readlines()

with open("data/raw/bible-uedin.am-en.en", "r", encoding="utf-8") as f:
    english_lines = f.readlines()

print(f"Total Amharic lines: {len(amharic_lines)}")
print(f"Total English lines: {len(english_lines)}")
print(f"Lines match: {len(amharic_lines) == len(english_lines)}")

print("\n--- First 5 parallel pairs ---")
for i in range(5):
    print(f"\nEN: {english_lines[i].strip()}")
    print(f"AM: {amharic_lines[i].strip()}")

print("\n--- Check for empty lines ---")
empty_am = sum(1 for line in amharic_lines if len(line.strip()) == 0)
empty_en = sum(1 for line in english_lines if len(line.strip()) == 0)
print(f"Empty Amharic lines: {empty_am}")
print(f"Empty English lines: {empty_en}")

print("\n--- Average line length ---")
avg_am = sum(len(l.strip()) for l in amharic_lines) / len(amharic_lines)
avg_en = sum(len(l.strip()) for l in english_lines) / len(english_lines)
print(f"Avg Amharic line length: {avg_am:.1f} chars")
print(f"Avg English line length: {avg_en:.1f} chars")
```
